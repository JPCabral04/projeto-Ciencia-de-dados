```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
import numpy as np
```

```python
df = pd.read_csv('diabetes_final.csv')

# features numéricas
numerical_features = df.columns.drop('Outcome')

X = df[numerical_features]
y = df['Outcome']

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

O PCA é muito sensível à escala dos números. O `RobustScaler` foi usado para colocar todas as variáveis na mesma escala, mas ignorando outliers, garantindo que o PCA foque na variação real dos dados e não em erros extremos.

```python
pca = PCA() 
X_pca = pca.fit_transform(X_scaled)

print("Variância explicada por componente:", pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=6, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Variância explicada por componente:", pca.explained_variance_ratio_)
print("Variância total:", pca.explained_variance_ratio_.sum())
```

Ao observar a "Variância Explicada Acumulada", nota-se que com 6 componentes você preserva cerca de 90% da informação original (variância). Isso reduz a complexidade do modelo (ruído) sem perder os padrões importantes dos dados.
